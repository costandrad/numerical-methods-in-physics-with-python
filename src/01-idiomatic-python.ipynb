## 1.5 Gráficos Básicos com `matplotlib`

```python
import matplotlib.pyplot as plt
```

```python
def plotex(cxs, cys, dxs, dys):
    plt.xlabel("$x$", fontsize=12)
    plt.ylabel("$f(x)$", fontsize=12)
    plt.plot(cxs, cys, 'r-', label="Modelo Teórico")
    plt.plot(dxs, dys, 'b--^', label="Dados experimentais")
    plt.legend()
    plt.plot()
```

```python
cxs = [0.1*i for i in range(60)]
cys = [x**2 for x in cxs]
dxs = [i for i in range(7)]
dys = [x**1.8 - 0.5 for x in dxs]

plotex(cxs, cys, dxs, dys)
```
